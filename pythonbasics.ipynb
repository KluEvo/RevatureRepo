```python

import os

bids = {}

def get_bid():
    global bids
    name = input("Name: ")
    bid = int(input("bid amount: $"))
    os.system('cls')
    bids[name] = bid

bidders = True
highest_bid = ["", 0]
while bidders:
    get_bid()
    more_bids = input("Any more bidders? (y/n): ")
    if more_bids.lower() == "n":
        break
print(bids)

for bidder, bid in bids.items():
    if bid > highest_bid[1]:
        highest_bid[1] = bid
        highest_bid[0] = bidder

print(f"{highest_bid[0]} won the auction with a bid of ${highest_bid[1]}")
```

```python
#Challenge: Clean data with Pandas  
import pandas as pd
import numpy as np
from book_generator_bad_data_service import generate_bad_books


generate_bad_books()

df = pd.read_json("books_dirty.json")
df = df.replace('', np.nan)

df['publication_year'] = pd.to_numeric(df['publication_year'], errors='coerce')

df['publication_year'] = df['publication_year'].clip(lower=int(1800), upper=int(2026))

df['page_count'] = pd.to_numeric(df['page_count'], errors='coerce')
df.loc[df['page_count'] <= 0, 'page_count'] = np.nan

df['average_rating'] = pd.to_numeric(df['average_rating'], errors='coerce')
df.loc[
    (df['average_rating'] < 0) | (df['average_rating'] > 5),
    'average_rating'
] = np.nan

df['price_usd'] = pd.to_numeric(df['price_usd'], errors='coerce')
df.loc[df['price_usd'] < 0, 'price_usd'] = np.nan

df['sales_millions'] = pd.to_numeric(df['sales_millions'], errors='coerce')
df.loc[df['sales_millions'] < 0, 'sales_millions'] = np.nan




df['last_checkout'] = pd.to_datetime(df['last_checkout'], errors='coerce')

bool_map = {
    'true': True,
    'false': False,
    True: True,
    False: False
}

df['in_print'] = df['in_print'].map(bool_map)
df['available'] = df['available'].map(bool_map)




df["genre"] = df["genre"].str.title()
df["language"] = df["language"].str.title()
df["format"] = df["format"].str.title()
df["publisher"] = df['publisher'].str.title()




df = df.drop_duplicates(subset=['title', 'author'], keep='first')

# only dropping on more critical fields because according to my calculations, 
# there is a < 0.00013 chance (< 0.013%) for any given book to have only values
# that will not be dropped via dropna. 
required_fields = ['title', 'author', 'genre', 'publication_year', 'page_count', 'language']
df = df.dropna(subset=required_fields)

# could drop all of them with
# df = df.dropna()


print(df.size)
print(df.head())
```
