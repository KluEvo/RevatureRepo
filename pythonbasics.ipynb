```python
raw_username = " Py_Thon-ROCKS123"

username = raw_username.strip().lower().replace("-", "_").replace("!", "")

has_digit = any(char.isdigit() for char in username)
contains_python = "python" in username
first_eight = username[0:8]
print(f"Username: {username}, contains a digit: {has_digit}, contains 'python': {contains_python}, first 8 chars: {first_eight}")
```

```python
students = [("Alice", 20, "A"), ("Bob", 19, "B"), ("Charlie", 21, "A"),("Diana", 20, "C")]
# Print only students with an 'A'
[print(student) for student in students if student[2] == "A"]
print("===")
# Print all names
[print(student[0]) for student in students]
print("===")
# Give all 20yo a 'B' then print list
print( [ tuple(["B" if x == student[2] else x for x in student]) if student[1] == 20 else student for student in students])
# alternative method
print( [ (name, age, "B") if age == 20 else (name, age, grade) for (name, age, grade) in students])
```

```python

import os

bids = {}

def get_bid():
    global bids
    name = input("Name: ")
    bid = int(input("bid amount: $"))
    os.system('cls')
    bids[name] = bid

bidders = True
highest_bid = ["", 0]
while bidders:
    get_bid()
    more_bids = input("Any more bidders? (y/n): ")
    if more_bids.lower() == "n":
        break
print(bids)

for bidder, bid in bids.items():
    if bid > highest_bid[1]:
        highest_bid[1] = bid
        highest_bid[0] = bidder

print(f"{highest_bid[0]} won the auction with a bid of ${highest_bid[1]}")
```
